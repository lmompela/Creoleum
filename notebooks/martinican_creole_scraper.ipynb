```python
import requests
from bs4 import BeautifulSoup
import argparse
import os
from os import listdir
from os.path import isfile, join
```

```python
def author_url_collector(file):
    print("Collecting Author's URLs...")
    author_url_dict = {}
    with open(file, encoding = "utf-8", mode = "r") as f:
        data = f.readlines()
        f.close()
        for line in data:
            line = line.split("\t")
            if len(line) > 1:
                author_url_dict.update({line[0]:line[1]})
    print("Done Collecting Author's URLS...")
    return author_url_dict
        

def find_all_pages(file):
    author_url_dict = author_url_collector(file)
    print("Collecting URL of All Pages per Author...")
    for author, url in author_url_dict.items():
        response = requests.get(
            url=url,
        )
    
        soup = BeautifulSoup(response.content, 'html.parser')
        url = url.split()
        last_page = soup.find('a', title = "Aller à la dernière page")
        author_url_dict.update({author:url[0]})
        author_url_dict[author] = [author_url_dict[author]]
        if last_page is not None:
            last_page = last_page['href'].split("=")
            for x in range(1, int(last_page[1])+1):
                more_url = url[0].split("=")
                more_page = more_url[0]+"="+str(x)
                author_url_dict[author].append(more_page)
        else:
            continue
    print("Done collecting URL of All Pages per Author...")
    return author_url_dict    


def collect_articles_url(file):
    pages_urls = find_all_pages(file)
    print("Collecting URL of All Articles in All Pages per Author...")
    article_url_dict = {}
    for key, value in pages_urls.items(): # For x in range(key)
        for x in range(len(value)):
            url = value[x]
            response = requests.get(
            url=url,
            )
    
            soup = BeautifulSoup(response.content, 'html.parser')
            author = key
            for x in range(len(soup.find_all('a', text = "Lire la suite"))):
                article_url = soup.find_all('a', text = "Lire la suite")[x]['href']
                if author not in article_url_dict:
                    article_url_dict.update({author:article_url})
                    article_url_dict[author] = [article_url_dict[author]]
                else:
                    article_url_dict[author].append(article_url)
    print("Done collecting URL of All Articles in All Pages per Author...")
    return article_url_dict
        
def scrapeArticle(file,domain):
    article_url_dict = collect_articles_url(file)
    print("Scraping All Articles...")
    for key in article_url_dict.keys():
        directory = key.replace(" ","_")
        path = "Martinican_creole/" + str(directory)
        os.makedirs(path, exist_ok=True)
        for y in range(len(article_url_dict[key])):
            url = domain+article_url_dict[key][y]
            print(url)
            OUT = open(str(path)+"/"+str(directory)+str(y)+'.txt', encoding="utf-8", mode='w')
            response = requests.get(
            url=url,
            )
    
            soup = BeautifulSoup(response.content, 'html.parser')
            column_text = soup.find('div',{'class' : 'bd-blog-1'})
            for x in range(len(column_text.find_all('div', {'class' : 'field-item even'}))):
                text = column_text.find_all('div', {'class' : 'field-item even'})[x].get_text()
                text = text.replace('\xa0','').replace('\n ', '\n')
                OUT.write(text)
                OUT.write("\n")
            OUT.close()
    print("Done scraping All Articles...")
            
def main():
    file = "sources_auteurs_martinique.txt"
    domain = "https://www.montraykreyol.org"
    scrapeArticle(file,domain)
    
main()
```
